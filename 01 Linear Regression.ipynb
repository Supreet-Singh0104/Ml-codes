```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def generateDataset(m):
    #To generate m random dataset
    X = np.random.randn(m)*10
    noise = np.random.randn(m) 
    y= 3*X + 1 + 5*noise
    return X,y
```

```python
X,y = generateDataset(100)
print(X.shape,y.shape)
```

```python
def plotdata(X,y,color="orange",title="Data"):
    sns.set(style="whitegrid") 
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.scatter(X,y,c=color)
    plt.show()
    
plotdata(X,y)
```

```python
def normaliseData(X):
    X = (X-X.mean())/X.std()
    return X
```

```python
plotdata(X,y)
X = normaliseData(X)
plotdata(X,y)
```

```python
def trainTestSplit(X,y,split=0.8):
    m=X.shape[0]
    data= np.zeros((m,2))
    data[:,0]=X
    data[:,1]=y
    
    np.random.shuffle(data) #shuffle the data of data the pair will be shifted randomly
    split=int(m*split) #split m rows in tain and test
    XT = data[:split,0] # suppose m=100 then 0-79 rows
    yT = data[:split,1] # then 80-99 rows
    Xt = data[split: ,0]
    yt = data[split: ,1]    
    return XT,yT,Xt,yt
```

```python
XT,yT,Xt,yt = trainTestSplit(X,y)
```

```python
print(XT.shape,yT.shape)
print(Xt.shape,yt.shape)
```

```python
plt.scatter(XT,yT,color='orange',label="Train Data")
plt.scatter(Xt,yt,color='blue',label="Test Data")
plt.title("Train-Test Data")
plt.legend()
plt.show()
     
```
