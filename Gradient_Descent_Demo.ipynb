```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=np.arange(10)
Y=(X-5)**2+3

print(X,Y)
```

```python
#plt.style.use("seaborn")
sns.set(style="whitegrid")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=0
lr=0.1

plt.plot(X,Y)
for i in range(30):
    grad = 2*(x-5)
    x = x-lr *grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)

plt.show()
```
